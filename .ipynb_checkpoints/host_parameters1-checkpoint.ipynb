```python
from openff.toolkit.topology import Molecule
from openff.toolkit.typing.engines.smirnoff import ForceField
from openff.toolkit.typing.engines.smirnoff import LibraryChargeHandler
import json
from rdkit import Chem
from IPython.display import Image
from openeye import oedepict
from openeye import oechem
```

```python
%%time
# Assign am1bcc partial charges to the host, track how much time it takes
offmol.assign_partial_charges(partial_charge_method='am1bcc',
                             use_conformers=offmol.conformers)

offmol.to_file('OA_charged.sdf', file_format='sdf')
```

```python
# load up the charged host from the previous step
offmol = Molecule.from_file('OA_charged.sdf')

sage = ForceField('openff-2.0.0.offxml')
# Modify FF. Apply charges from the molecule 
new_librarycharge = sage['LibraryCharges'].LibraryChargeType.from_molecule(offmol)
sage['LibraryCharges'].parameters.append(new_librarycharge)
sage['LibraryCharges'].parameters
```

```python
sage.create_openmm_system(offmol.to_topology())
```

```python
qmol = oechem.OEQMol()
num = 2
print(list(data.keys())[num])
oechem.OEParseSmarts(qmol, list(data.keys())[num])

oemol = qmol
width=500
height=500
show_all_hydrogens=True
opts = oedepict.OE2DMolDisplayOptions(width, height, oedepict.OEScale_AutoScale)

if show_all_hydrogens:
    opts.SetHydrogenStyle(oedepict.OEHydrogenStyle_ImplicitAll)

oedepict.OEPrepareDepiction(oemol)
img = oedepict.OEImage(width, height)
display = oedepict.OE2DMolDisplay(oemol, opts)
oedepict.OERenderMolecule(img, display)
png = oedepict.OEWriteImageToString("png", img)
Image(png)
```
